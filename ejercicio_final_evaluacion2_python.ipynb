```python
import requests
import pandas as pd
```

```python
url ='https://beta.adalab.es/resources/apis/pelis/pelis.json'

llamada = requests.get(url)

if llamada.status_code == 200:
    datos_llamada = llamada.json()
    print("exito")
else:
    print("error")
```

```python
datos_llamada =llamada.json()
```

```python
for peli in datos_llamada:
    titulo = peli.get("titulo",None) # uso None que es igual a Null en sql
    año = peli.get("año","Null")  #.get para para evitar errores si falta algun campo
    genero =peli.get("genero",None)
    duracion = peli.get("duracion",None)
    contenido_adul_pelis = peli.get("adultos",None)
    print(f"Título: {titulo} | Género: {genero} | Duración: {duracion} min | Contenido adultos: {contenido_adul_pelis}")
```

```python
datos_tabla_pelis = []

for datos in datos_llamada:
    tupla = (        #tupla para que sql lo entienda
        datos["titulo"],
        datos["año"],
        datos["duracion"],
        datos["genero"],
        int(datos["adultos"])  # convierte booleano a 1/0
    )
    datos_tabla_pelis.append(tupla)

print(datos_tabla_pelis)

```

```python
listas_subtitulos = [] #la ultima tabla que es fk solo necesita esto dos parametros

for peli in datos_pelis:
    listas_subtitulos.append(f"{{'id':{peli[0]},'subtitulos':{peli[6]}}}") #asi dejo el diccionario para extraer los subtitulos

print(listas_subtitulos)

```

```python
# lista de películas
peliculas = listas_subtitulos
  
def generar_insert_subtitulos(peliculas):
   insert_values =[]
   for peli in peliculas:
       id_pelis = peli["id"]
       for idioma in peli["subtitulos"]:
           insert_values.append(f"({id_pelis},'{idioma}')")

# Crear la sentencia completa para la tabla temporal
sql = "INSERT INTO temp_subtitulos (id_pelis,codigo_api) VALUES\n" + ",\n".join(insert_values) + ";"

print(sql)
```
